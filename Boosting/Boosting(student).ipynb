```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostRegressor,GradientBoostingRegressor
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
df = pd.read_csv('student-mat.csv')
```

```python
df = df.drop(["G1","G2"], axis=1)
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
X = df.drop("G3", axis=1)
y = df["G3"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
model1 = AdaBoostRegressor(n_estimators=100,random_state=42)
model1.fit(X_train,y_train)
y_pred1 = model1.predict(X_test)
print("MSE:",mean_squared_error(y_test,y_pred1))
print("MAE:",mean_absolute_error(y_test,y_pred1))
print('r2_score:',r2_score(y_test,y_pred1))
```

```python
model2 = GradientBoostingRegressor(n_estimators=500,learning_rate=0.05,max_depth=3)
model2.fit(X_train,y_train)
y_pred2 = model2.predict(X_test)
print("MSE:",mean_squared_error(y_test,y_pred2))
print("MAE:",mean_absolute_error(y_test,y_pred2))
print('r2_score:',r2_score(y_test,y_pred2))
```

```python
model3 = XGBRegressor()
model3.fit(X_train,y_train)
y_pred3 = model3.predict(X_test)
print("MSE:",mean_squared_error(y_test,y_pred3))
print("MAE:",mean_absolute_error(y_test,y_pred3))
print('r2_score:',r2_score(y_test,y_pred3))
```

```python
model4 = CatBoostRegressor()
model4.fit(X_train,y_train)
y_pred4 = model4.predict(X_test)
print("MSE:",mean_squared_error(y_test,y_pred4))
print("MAE:",mean_absolute_error(y_test,y_pred4))
print('r2_score:',r2_score(y_test,y_pred4))
```

```python
model5 = LGBMRegressor()
model5.fit(X_train,y_train)
y_pred5 = model5.predict(X_test)
print("MSE:",mean_squared_error(y_test,y_pred5))
print("MAE:",mean_absolute_error(y_test,y_pred5))
print('r2_score:',r2_score(y_test,y_pred5))
```

```python
from sklearn.ensemble import RandomForestRegressor
model6 = RandomForestRegressor(n_estimators=200,max_depth=10,random_state=42)
model6.fit(X_train,y_train)
y_pred6 = model6.predict(X_test)
print("MSE:",mean_squared_error(y_test,y_pred6))
print("MAE:",mean_absolute_error(y_test,y_pred6))
print('r2_score:',r2_score(y_test,y_pred6))
```

```python
from sklearn.tree import DecisionTreeRegressor
model7=DecisionTreeRegressor(random_state=42,max_depth=6)
model7.fit(X_train,y_train)
y_pred7 = model7.predict(X_test)
print("MSE:",mean_squared_error(y_test,y_pred7))
print("MAE:",mean_absolute_error(y_test,y_pred7))
print('r2_score:',r2_score(y_test,y_pred7))
```
