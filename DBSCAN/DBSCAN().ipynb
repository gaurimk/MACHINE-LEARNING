```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
X = np.array([
    [1,2],[2,2],[2,3],[3,3],
    [8,7],[8,8],[7,8],
    [20,20]
])
```

```python
model = DBSCAN(eps=1.5, min_samples=2)
clusters = model.fit_predict(X)
print(clusters)
```

```python
plt.scatter(X[:,0], X[:,1], c=clusters, cmap="rainbow")
plt.title("DBSCAN Clustering")
plt.show()
```
